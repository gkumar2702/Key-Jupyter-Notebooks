```python
import json
import requests
import pandas as pd
import numpy as np
from io import BytesIO
from zipfile import ZipFile
from urllib.request import urlopen
from pandas.io.json import json_normalize

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 2000)
pd.options.display.float_format = '{:.2f}'.format
```

```python
def hazard_data(start, end):
    '''
    
    
    '''
    
    key = 'e721181f854ac2268ee8'
    start = pd.to_datetime(start,format='%Y-%m-%d')
    end = pd.to_datetime(end, format='%Y-%m-%d')
    
    start = start.strftime('%Y-%m-%dT%H:%M:%S')
    end = end.strftime('%Y-%m-%dT%H:%M:%S')
    
    hazard_ = pd.DataFrame()
    link = 'https://api.weathersource.com/v1/'+key+'/storms.json?timestamp_between='+start+','+end
    print(link)
    response = requests.get(link)
    json_obj = json.loads(response.content.decode('utf-8'))
    hazard_ = json_normalize(json_obj)
    
    return hazard_
```

```python
start_date = "2019-07-09"
end_date = "2020-06-15"

hazard_data = hazard_data(start_date, end_date)
```

```python
def storm_info(storm_id):
    '''
    
    
    '''
    
    key = 'e721181f854ac2268ee8'
    
    storm_ = pd.DataFrame()
    link = 'https://api.weathersource.com/v1/'+key+'/storms/'+storm_id+'/hazards/hurricanes.geojson.zip'
    print(link)
    response = requests.get(link)
    json_obj = json.loads(response).json()
    data = pd.read_json(json_obj)
    storm_ = data['features']
    
    return data
```
