```python
df_markers = pd.read_csv('gs://aes-analytics-0002-curated/Outage_Restoration/Live_Data_Curation/Mapping_Tables/IPL_OMS_MARKER_MAPPING.csv', index_col=0)
df_markers.head()
```

```python
import folium

#create a map
this_map = folium.Map(prefer_canvas=True)

def plotDot(point):
    '''input: series that contains a numeric named latitude and a numeric named longitude
    this function creates a CircleMarker and adds it to your this_map'''
    folium.CircleMarker(location=[point.Latitude, point.Longitude],
                        radius=5,
                        weight=5).add_child(folium.Popup(point.Marker)).add_to(this_map)

#use df.apply(,axis=1) to "iterate" through every row in your dataframe
df_markers.apply(plotDot, axis = 1)


#Set the zoom to the maximum possible
this_map.fit_bounds(this_map.get_bounds())
this_map
```

```python
Region = ['34th', '34th', '34th', 'ARL', 'ARL', 'MILL', 'MILL', 'ENGLISH', 'ENGLISH', 'MILL',
          'MILL', 'ENGLISH', 'ENGLISH', 'W.I.', 'W.I.', 'SOUTH', 'SOUTH', 'W.I.', 'W.I.', 'W.I.']

df_markers['Region'] = Region
df_markers.head()
```

```python
#create a map
this_map = folium.Map(prefer_canvas=True)

def plotDot(point):
    '''input: series that contains a numeric named latitude and a numeric named longitude
    this function creates a CircleMarker and adds it to your this_map'''
    if point.Region == '34th':
        folium.CircleMarker(location=[point.Latitude, point.Longitude],
                            radius=5,
                            weight=5, color='blue').add_child(folium.Popup(point.Marker)).add_to(this_map)
    elif point.Region == 'ARL':
        folium.CircleMarker(location=[point.Latitude, point.Longitude],
                            radius=5,
                            weight=5, color='red').add_child(folium.Popup(point.Marker)).add_to(this_map)
    elif point.Region == 'MILL':
        folium.CircleMarker(location=[point.Latitude, point.Longitude],
                            radius=5,
                            weight=5, color='green').add_child(folium.Popup(point.Marker)).add_to(this_map)
    elif point.Region == 'ENGLISH':
        folium.CircleMarker(location=[point.Latitude, point.Longitude],
                            radius=5,
                            weight=5, color='black').add_child(folium.Popup(point.Marker)).add_to(this_map)
    elif point.Region == 'W.I.':
        folium.CircleMarker(location=[point.Latitude, point.Longitude],
                            radius=5,
                            weight=5, color='grey').add_child(folium.Popup(point.Marker)).add_to(this_map)
    else:
        folium.CircleMarker(location=[point.Latitude, point.Longitude],
                            radius=5,
                            weight=5, color='orange').add_child(folium.Popup(point.Marker)).add_to(this_map)

#use df.apply(,axis=1) to "iterate" through every row in your dataframe
df_markers.apply(plotDot, axis = 1)


#Set the zoom to the maximum possible
this_map.fit_bounds(this_map.get_bounds())
this_map
```
